```python
#SMOTE- Balancing Dataset

from sklearn.utils import resample
# Find the class with the most occurrences (majority class)
majority_class = df['isFraud'].value_counts().idxmax()

# Split the dataset by class labels
df_majority = df[df['isFraud'] == majority_class]
df_minority = df[df['isFraud'] != majority_class]

# Perform upsampling for minority classes
upsampled = pd.DataFrame()

# For each minority class, resample it to match the size of the majority class
for class_label in df_minority['isFraud'].unique():
    # Extract samples for each class
    df_class = df_minority[df_minority['isFraud'] == class_label]
    
    # Upsample minority class to the size of the majority class
    df_class_upsampled = resample(df_class, 
                                  replace=True,  # With replacement
                                  n_samples=len(df_majority),  # Match number of majority class samples
                                  random_state=42)
    
    # Append the upsampled class to the final dataset
    upsampled = pd.concat([upsampled,df_class_upsampled])

# Combine the upsampled dataset with the majority class to create the balanced dataset
df_balanced = pd.concat([df_majority,upsampled])
```

CLUSTERING USING K-MEANS AND ELBOW METHOD

```python
from sklearn.cluster import KMeans
```

```python
def elbow_method(X, max_clusters=10):
    """
    Uses the elbow method to find the optimal number of clusters.

    Parameters:
    - X: array-like, shape (n_samples, n_features)
    - max_clusters: int, maximum number of clusters to test

    Returns:
    - inertias: list of inertia values for each cluster count
    """
    inertias = []
    cluster_range = range(1, max_clusters + 1)
    
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)
    
    # Plot the Elbow Graph
    plt.figure(figsize=(8, 5))
    plt.plot(cluster_range, inertias, marker='o', linestyle='--', color='b')
    plt.title('Elbow Method to Determine Optimal K')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.xticks(cluster_range)
    plt.grid()
    plt.show()

    return inertias

# Run the elbow method
elbow_method(X)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt



# Step 2: Preprocessing
# Check for missing values
print("Missing values per column:")
print(X.isnull().sum())

# Fill or drop missing values if any
X.fillna(0, inplace=True)  # Replace NaN with 0 (or use other strategies)

# Select relevant columns for clustering
features = ['step', 'type', 'amount', 'newbalanceOrig', 'newbalanceDest']
X = X[features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-means clustering
# Choose the number of clusters (k)
k = 3  # Adjust this value based on your analysis

kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# Add cluster labels to the original DataFrame
X['Cluster'] = kmeans.labels_

# Step 4: Analyze the clusters
print("Cluster Centers:")
print(kmeans.cluster_centers_)

print("Cluster Labels Assigned to Data:")
print(X['Cluster'].value_counts())

# Step 5: Visualize the clusters (optional)
# Reduce dimensionality to 2D for visualization using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=X['Cluster'], cmap='viridis', s=10)
plt.title("K-means Clustering Visualization")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.show()
```
